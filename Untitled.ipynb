```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Step 1: Fetch stock data
def get_stock_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data
```

```python
# Fetch stock data
ticker = 'AAPL'  # Example: Apple stock
start_date = '2020-01-01'
end_date = '2023-01-01'

stock_data = get_stock_data(ticker, start_date, end_date)

# Display summary statistics in a better format
summary_statistics = stock_data.describe()
summary_statistics_df = pd.DataFrame(summary_statistics)  # Convert to DataFrame
display(summary_statistics_df)  # Use display() to show the DataFrame
```

```python
# Step 2: Preprocess data for machine learning
def preprocess_data(stock_data):
    stock_data['Prediction'] = stock_data['Close'].shift(-1)  # Predict the next day's close price
    X = stock_data[['Open', 'Close', 'Volume']].values[:-1]  # Features: Open, Close, Volume
    y = stock_data['Prediction'].values[:-1]  # Labels: Next day close prices
    return X, y

# Preprocess data
X, y = preprocess_data(stock_data)
```

```python
# Step 3: Train-test split
def train_test_split_data(X, y, test_size=0.2):
    return train_test_split(X, y, test_size=test_size, random_state=42)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split_data(X, y, test_size=0.2)
```

```python

# Step 4: Train Linear Regression model
def train_model(X_train, y_train):
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model

# Train model
model = train_model(X_train, y_train)
```

```python

# Step 5: Make predictions and calculate error
def make_predictions(model, X_test):
    predictions = model.predict(X_test)
    return predictions

# Make predictions
predictions = make_predictions(model, X_test)
```

```python
# Calculate error
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
# Step 6: Visualize the results
def plot_predictions(stock_data, predictions):
    valid = stock_data[len(stock_data) - len(predictions):].copy()  # Create a copy of the validation set
    valid.loc[:, 'Predictions'] = predictions  # Assign predictions to valid DataFrame

    plt.figure(figsize=(16,8))
    plt.title('Stock Price Prediction')
    plt.plot(stock_data['Close'], label='Actual Price')  # Plot actual close prices
    plt.plot(valid.index, valid['Predictions'], label='Predicted Price', color='orange')  # Plot predicted prices
    plt.legend()
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()
```

```python
# Plot predictions
plot_predictions(stock_data, predictions)
```

```python
# Import necessary libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential

# Fetch stock data
start = '2010-01-01'
end = '2020-01-01'
df = yf.download('AAPL', start=start, end=end)
print(df.head())

# Plotting the Open and Close prices
plt.figure(figsize=(12,6))
plt.plot(df['Open'], label='Open Price', color='blue')
plt.plot(df['Close'], label='Close Price', color='orange')
plt.title('Open and Close Prices of AAPL')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Dropping unnecessary columns
df = df.drop(['Adj Close'], axis=1)
print(df.tail())

# Plotting the Close price
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close Price')
plt.title('Close Price of AAPL')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Calculating moving averages
ma100 = df['Close'].rolling(100).mean()
ma200 = df['Close'].rolling(200).mean()

# Plotting moving averages
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close Price')
plt.plot(ma100, 'r', label='100-Day MA')
plt.plot(ma200, 'g', label='200-Day MA')
plt.title('Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Splitting Data into Training and Testing
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
print(f'Training Data Shape: {data_training.shape}')
print(f'Testing Data Shape: {data_testing.shape}')

# Preprocessing the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_training_array = scaler.fit_transform(data_training)

# Creating the training dataset
x_train = []
y_train = []
for i in range(100, data_training_array.shape[0]): 
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)

# Building the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(units=80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=120, activation='relu', return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(units=1))

# Model summary
model.summary()

# Compiling and training the model
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(x_train, y_train, epochs=50)

# Save the model
model.save('keras.h5')

# Testing the model
data_testing.head()

# Preparing the final dataset for testing
past_100_days = data_training.tail(100)
final_df = pd.concat([past_100_days, data_testing], ignore_index=True)

input_data = scaler.fit_transform(final_df)

# Creating the testing dataset
x_test = []
y_test = []
for i in range(100, input_data.shape[0]): 
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test = np.array(x_test)

# Making predictions
y_predicted = model.predict(x_test)

# Inverse scaling the predictions
scale_factor = 1 / scaler.scale_[0]  # The scale factor for inverse transformation
y_predicted = y_predicted * scale_factor
y_test = np.array(y_test) * scale_factor

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Original Price', color='blue')
plt.plot(y_predicted, label='Predicted Price', color='red')
plt.title('Original vs Predicted Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
